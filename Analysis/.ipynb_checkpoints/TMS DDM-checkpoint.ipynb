## Functions

```python
# plotting function
def plot_mod(model, node1, node2, param, title):
    n1, n2 = model.nodes_db.node[[node1,node2]]
    hddm.analyze.plot_posterior_nodes([n1, n2])
    plt.xlabel(param)
    plt.ylabel('Posterior probability')
    plt.title(title)
    
    print("P(node1 > node2) = ", (n1.trace() > n2.trace()).mean())
    print("P(node2 > node1) = ", (n2.trace() > n1.trace()).mean())
```

### full_m1: v ~ 0 + C(trial_type)\*C(site_loc), a ~ 0 + C(trial_type)\*C(site_loc), t ~ 0 + C(trial_type)*C(site_loc)

```python
# same model as above - different way of doing it

model3 = hddm.HDDMRegressor(data, {"v ~ 0 + C(trial_type)*C(site_loc)",\
                                   "a ~ 0 + C(trial_type)*C(site_loc)",\
                                   "t ~ 0 + C(trial_type)*C(site_loc)"}, p_outlier=.05)
model3.sample(2000, burn=200)
```

```python
t_hard, t_easy = model3.nodes_db.node[["t_C(trial_type)[hard]","t_C(trial_type)[easy]"]]
hddm.analyze.plot_posterior_nodes([t_hard, t_easy])
plt.xlabel('t - easy vs hard')
plt.ylabel('Posterior probability')
plt.title('Posterior of bias group means')
```

```python
v_hard, v_easy = model3.nodes_db.node[["v_C(trial_type)[hard]","v_C(trial_type)[easy]"]]
hddm.analyze.plot_posterior_nodes([v_hard, v_easy])
plt.xlabel('v - easy vs hard')
plt.ylabel('Posterior probability')
plt.title('Posterior of bias group means')
```

```python
a_hard, a_easy = model3.nodes_db.node[["a_C(trial_type)[hard]","a_C(trial_type)[easy]"]]
hddm.analyze.plot_posterior_nodes([a_hard, a_easy])
plt.xlabel('a - easy vs hard')
plt.ylabel('Posterior probability')
plt.title('Posterior of bias group means')
```

### full_m2: v ~ C(trial_type)\*C(site_loc), a ~ C(trial_type)\*C(site_loc), t ~ C(trial_type)*C(site_loc)

```python
full_m2 = hddm.HDDMRegressor(data, {"v ~ C(trial_type)*C(site_loc)",\
                                   "a ~ C(trial_type)*C(site_loc)",\
                                   "t ~ C(trial_type)*C(site_loc)"}, p_outlier=.05)
full_m2.sample(3000, burn=200, dbname=path2mods + 'full_m2.db', db='pickle')
full_m2.save(path2mods + 'full_m2')
# to load: full_m2 = hddm.load(path2mods + 'full_m2')
```

```python
# get stats for each condition and save as csv 
full_m2_df = full_m2.gen_stats().reset_index() 
full_m2_df = full_m2_df.rename(columns={'index': 'condition'})
full_m2_df.to_csv(path2mods + 'full_m2_genstats.csv')
full_m2_df.head()
```

```python
full_m2_df
```

v_C(trial_type)[T.hard] - main effect of trial type  
v_C(site_loc)[T.dACC] - main effect of site  
v_C(trial_type)[T.hard]:C(site_loc)[T.dACC] - interaction term  

```python
# Drift rate fig & stats
v_trial_type = full_m2.nodes_db.node["v_C(trial_type)[T.hard]"]
hddm.analyze.plot_posterior_nodes([v_trial_type], bins=20)
plt.xlabel('Regression coefficient for trial type')
plt.title('Main effect of trial type (easy/hard) on drift rate (v)',fontweight="bold")

print("\033[1m Main effect of trial type (easy/hard) on drift rate (v): \033[0m")
print("P(v_trial_type < 0) = ", (v_trial_type.trace() < 0).mean())
print("P(v_trial_type > 0) = ", (v_trial_type.trace() > 0).mean())
print("\n")

# Threshold fig & stats
a_trial_type = full_m2.nodes_db.node["a_C(trial_type)[T.hard]"]
hddm.analyze.plot_posterior_nodes([a_trial_type], bins=20)
plt.xlabel('a_trial_type')
plt.title('Main effect of trial type (easy/hard) on threshold',fontweight="bold")

print("\033[1m Main effect of trial type (easy/hard) on threshold (a): \033[0m")
print("P(a_trial_type < 0) = ", (a_trial_type.trace() < 0).mean())
print("P(a_trial_type > 0) = ", (a_trial_type.trace() > 0).mean())
print("\n")

# Non-decision time fig & stats
t_trial_type = full_m2.nodes_db.node["t_C(trial_type)[T.hard]"]
hddm.analyze.plot_posterior_nodes([t_trial_type], bins=20)
plt.xlabel('t_trial_type')
plt.title('Main effect of trial type (easy/hard) on non-decision time',fontweight="bold")

print("\033[1m Main effect of trial type (easy/hard) on non-decision time (t): \033[0m")
print("P(t_trial_type < 0) = ", (t_trial_type.trace() < 0).mean())
print("P(t_trial_type > 0) = ", (t_trial_type.trace() > 0).mean())
print("\n")
```

```python
all_idp = hddm.HDDMRegressor(data, {"v ~ IDP",\
                                    "a ~ IDP",\
                                    "t ~ IDP"}, p_outlier=.05)
all_idp.sample(1000, burn=100, dbname=path2mods + 'all_idp.db', db='pickle')
all_idp.save(path2mods + 'all_idp')
# to load: all_idp = hddm.load(path2mods + 'all_idp')
```

```python
a_IDP = all_idp.nodes_db.node["a_IDP"]
hddm.analyze.plot_posterior_nodes([a_IDP], bins=20)
plt.xlabel('a_IDP')
print("P(a_IDP < 0) = ", (a_IDP.trace() < 0).mean())
print("P(a_IDP > 0) = ", (a_IDP.trace() > 0).mean())
```

```python
v_IDP = all_idp.nodes_db.node["v_IDP"]
hddm.analyze.plot_posterior_nodes([v_IDP], bins=20)
plt.xlabel('v_IDP')
print("P(v_IDP < 0) = ", (v_IDP.trace() < 0).mean())
print("P(v_IDP > 0) = ", (v_IDP.trace() > 0).mean())
```

```python
t_IDP = all_idp.nodes_db.node["t_IDP"]
hddm.analyze.plot_posterior_nodes([t_IDP], bins=20)
plt.xlabel('t_IDP')
print("P(t_IDP < 0) = ", (t_IDP.trace() < 0).mean())
print("P(t_IDP > 0) = ", (t_IDP.trace() > 0).mean())
```
